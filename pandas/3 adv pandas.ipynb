```python
# ------convert uro into rs ,then new col fare in inr
# apply fun work on entire col

df['Fare_Inr']= df["Fare"].apply(lambda x:x*80)
```

```python
df.head()
```

```python
df["Name_len"]= df['Name'].apply(len)
```

```python
df.head()
```

```python
df['Desription of sex']=df['Sex'].apply(lambda a:'This is boy or men' if a=='male' else'This is girl or women')
df.head(4)
```

```python
def cat_fare(x):
    if x<10:
         return"cheap"
    elif x>=10 and x<20:
        return 'mid'
    else :
         return 'high'
        
df['cat_fare'] = df['Fare'].apply(cat_fare)
df.head(2)
```

```python
df['Pclass_name'] = df['Pclass'].apply(lambda b: 'First class' if b == 3 else ('Second class' if b > 3 else 'Third class'))
```
